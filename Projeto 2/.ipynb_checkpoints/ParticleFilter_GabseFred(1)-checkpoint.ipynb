4. Crie uma lista para simular os movimentos do robô. Veja a lista chamada <code>movimentos</code> que é usada na seção *Teste de animação*, mais abaixo

6. Desenhe as partículas após o deslocamento

7. Descubra como calcular $P(D|H)$ analisando o Python Notebook. A função que traça os raios está no modulo inspercles

Programe a aplicação da equação que está na pág. 853 do livro do Norvig $$P(z_t | x_t) = \alpha \prod_{j}^M{e^{\frac{-(z_j - \hat{z_j})}{2\sigma^2}}}$$

Ou seja, para cada raio estimado $\hat{z_j}$ e leitura real $z_j$ calcule a probabilidade daquele raio ter acontecido dado que a posição do robô fosse de fato a da partícula


8 . Usando os resultados de 7 calcule $P(H|D)$ e insira numa lista de probabilidades das partículas

9 . Reamostre as partículas de forma proporcional à probabilidade

```python
plt.ioff() # Desliga o modo interativo, para nao aparecerem muitas imagens no meio

frames = 1
prob_list = []
robot_lidar_read = []
particle_lidar_read = []


# Pra cada movimento
for delta in movimentos:
    
    # Move o robo
    for i in range(len(pose)):
        pose[i]+=delta[i]
        
    print "================ ROBO SE MEXEU " + str(frames) + " ================="
    
    # Simula o lidar do robo
    leituras_robo, lidar_map = inspercles.nb_simulate_lidar(pose, angles, np_image)   
#   leituras_robo, lidar_map = inspercles.nb_simulate_lidar_fast(pose, angles, np_image

    print "\n LEITURAS ROBO \n"
    print leituras_robo
    soma = 0
    somaprob = 0
    pl = []

#     robot_lidar_read.append(leituras)

    print "LEITURAS PARTICULAS \n"   
    
    # Para cada uma das n particulas
    for j in range(len(particulas)):
        particulas[j].x += delta[0] + np.random.normal(0,2)
        particulas[j].y += delta[1] + np.random.normal(0,2)
        particulas[j].theta += delta[2] + np.random.normal(0,5)
            
        # Simula o lidar das particulas
        leituras_particulas, lidar_map2 = inspercles.nb_simulate_lidar((particulas[j].x,particulas[j].y,particulas[j].theta),angles,np_image)
#       leituras_particulas, lidar_map = inspercles.nb_simulate_lidar_fast((particulas[j].x,particulas[j].y,particulas[j].theta),angles,np_image,retorno_lidar_particulas,output_image=False)
        
        # Lista com as chaves do dicionario do lidar - relacao raio//distancia
        keys = leituras_particulas.keys()
        
        #Produto inicial = 1 p/ cada particula
        produto = 1 
        
        # Para cada uma das chaves
        for key in keys:
            # Define as variaveis dos raios
            raio_particula = round(leituras_particulas[key],3)
            raio_robo = round(leituras_robo[key],3)
            # Faz a conta
            exp = math.e**(-(raio_robo - raio_particula)/8)
            # E multiplica no produto
            produto *= exp            
            #print("part:",raio_particula," robo:",raio_robo)
            
#         print '\n'
        # coloca o produto final da particula na lista pl
        pl.append(produto)
    
    
    soma = sum(pl)
    alpha = 1/soma     
    newprob = []
    
    for prob in pl:
        newprob.append(alpha*prob)
        
    print'\n'
    print ('newprob:',newprob)
        
    for j in range(len(particulas)):
        particulas[j].w = newprob[j]
    
    
    novas_particulas = Particle.draw_random_sample(particulas, newprob, 50)
#     print("PESOMENkk:",particulas.w)
        
#         print(particulas[j].w)
    

    # Desenha as particulas
    ax = inspercles.nb_draw_map(mapa_png, pose=pose, robot=True, particles=particulas)
    #ax.imshow(occupancy_image, alph.2)
    
    std_lidar = 2
    
    # Desenha o mapa do lidar
    ax.imshow(lidar_map, alpha=0.5)
    
    plt.savefig("nosso%04d.png"%frames, bounds="tight")
    
    frames+=1
    plt.close('all')
    
    particulas = novas_particulas


plt.ion()
```
